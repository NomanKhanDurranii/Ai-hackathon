```python
import csv
import time
import re
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
```

```python
driver = webdriver.Chrome()
driver.get("https://www.tradingview.com/markets/stocks-usa/market-movers-all-stocks/")

```

```python
CompleteTable = driver.find_element(By.XPATH , "/html/body/div[2]/main/div/div[2]/div/div[4]/div/div[2]/div[2]/div/div/table/tbody")
rows = CompleteTable.find_elements(By.CSS_SELECTOR , 'tr')

with open("symbol_link.csv", "a") as file:
    file.write("Exchange , Symbol\n")

for i in rows:
    attribute = i.get_attribute('data-rowkey')
    both = attribute.split(":")
    with open("symbol_link.csv", "a") as file:
        file.write(f"{both[0]} , {both[1]}\n")
    print(both)
```
